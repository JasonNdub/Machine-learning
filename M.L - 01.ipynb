# Bank.csv_(Dataset)

```python
import numpy as np # import numpy
import pandas as pd # import pandas
```

```python
df = pd.read_csv('bank.csv', sep = ';') # import the dataset (bank.csv)
# parameter "sep = ';' " tells the "read_csv" function that the columns in the "CSV" file are separated by a ";" instead of the default ","
```

```python
df.head(10) # print 0 to 9 lines
```

```python
def replace_marital (val):   # This function checks 
    if val == "single":       # if the value 'val' is single
        return 0             # if it is single it returns '0'
    else:                    # if it is not single
        return 1             # it returns '1'
df["marital"] = df["marital"].apply(replace_marital)
# This applies the replace_marital function to each value in the "marital" column.
```

```python
df.head()
```

```python
# This line of code maps the values in the "housing" column of the DataFrame df to numerical values. Specifically:
df ["housing"] =df["housing"].map({
    "no" : 0, # "no" is mapped to 0
    "yes": 1  # "yes" is mapped to 1
}.get)        # The .map() function is used to apply this mapping,
              # and the .get method ensures that any value not specified in the dictionary will be returned as None
```

```python
df.head()
```

```python
df ["loan"] =df["loan"].map({  # same as 92 no. para
    "no" : 0,
    "yes": 1
}.get) 
```

```python
df.head()
```

```python
df["job"].unique() # It shows all unique values in the "job" column.
```

```python
# This code replaces the values in the "job" column of the DataFrame df with corresponding numerical codes.
df["job"].replace ( {
   'unemployed' :0,  # 'unemployed' becomes 0
    'services' :1    # 'services' becomes 1
    'management' :2, # 'management' becomes 2 
    'blue-collar' :3,
    'self-employed' :4,
    'technician' :5,
    'entrepreneur' :6,
    'admin.' :7,
    'student' :8,
    'housemaid' :9,
    'retired' :10,
    'unknown' :np.nan, # If the value is 'unknown', it is replaced with NaN (missing value) 
}, inplace= True)  # The inplace=True argument updates the DataFrame directly.
```

```python
df["job"].unique() # It shows all unique values in the "job" column.
```

Normalizes the "balance" column by scaling the values to a range between 0 and 1.

```python
df["balance"] = df["balance"].apply(lambda v:(v- df["balance"].min()) / (df["balance"].max () - df["balance"].min ()))
```
